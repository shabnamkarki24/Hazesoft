## finding out top 5 highest rates movies

```python
import pandas as pd
df=pd.read_csv("http://bit.ly/imdbratings")
top5=df.sort_values(by="star_rating", ascending=False).head(5)
top5

```

```python
shortest_movie= df.loc[df['duration'].idxmin(),['title', 'duration', 'genre']]
longest_movie=df.loc[df['duration'].idxmax(), ['title', 'duration', 'genre']]
print("shortest movie:\n", shortest_movie)
print("longest movie:\n", longest_movie)
    
```
